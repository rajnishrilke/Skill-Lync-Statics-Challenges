```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(insurance["sex"], insurance["smoker"])

# Calculate the chi-square statistic
chi_square_statistic, p_value, df, _ = chi2_contingency(contingency_table)

# Print the results
print("Chi-square statistic:", chi_square_statistic)
print("P-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("The variables are statistically dependent.")
else:
    print("The variables are statistically independent.")
```
