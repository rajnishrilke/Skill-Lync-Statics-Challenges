## Do wines with higher alcohol content receive better ratings?

```python
columns_to_check = ['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','density','pH','sulphates','alcohol','quality']
# Calculate the Z-scores for each data point in the specified columns
z_scores = np.abs((df_wine[columns_to_check] - df_wine[columns_to_check].mean()) / df_wine[columns_to_check].std())

# Set the threshold for outlier detection (Z-score greater than 3)
threshold = 3

# Remove the rows with outliers
df_wine[(z_scores < threshold)]

# Display the dataset without outliers
#dataset_no_outliers
```

## Find how many variables are independent.

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df_wine["citric acid"], df_wine["pH"])

# Calculate the chi-square statistic
chi_square_statistic, p_value, df, _ = chi2_contingency(contingency_table)

# Print the results
print("Chi-square statistic:", chi_square_statistic)
print("P-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("The variables are statistically dependent.")
else:
    print("The variables are statistically independent.")
```

## Create a new column as the acidity level. Divide the existing data into 5 groups based on the acidity level. The levels are to be named ['High', 'Moderately_High', 'Medium', 'Low']. Acidity may be obtained from the PH.

```python
bin_edges = [2.72,3.11,3.21,3.32,4.01] 
bin_names = ['very_high','high','medium','low'] 
df_wine['acidity_levels'] = pd.cut(df_wine['pH'], bin_edges, labels=bin_names)
df_wine.head()
```

## Find the mean quality of each acidity level with groupby. Use groupby

```python
df_wine.groupby('acidity_levels')['quality'].mean()
```

## Is the density of red wine and white wine similar?

```python
from scipy import stats
# Extract the density for red and white wines
red_wine_density = df_wine.loc[df_wine['color'] == 'red', 'density']
white_wine_density = df_wine.loc[df_wine['color'] == 'white', 'density']

# Perform a t-test
t_statistic, p_value = stats.ttest_ind(red_wine_density, white_wine_density, equal_var=False)

# Compare the p-value to the significance level (e.g., 0.05)
alpha = 0.05
if p_value < alpha:
    print("The density of red wine and white wine is significantly different.")
else:
    print("The density of red wine and white wine is not significantly different.")
```
